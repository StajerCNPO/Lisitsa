```python
import matplotlib.pyplot as plt
from venn import venn
```

```python
a  = [7,5,6,2,9,18]
b = [5,8,14, 64, 4, 7]
c = [5,8]
d = [11,1,5]
set_a = set(a)
set_b = set(b)
set_c = set(c)
set_d = set(d)
```

```python
union_a_b = set.union(set_a, set_b)
print(union_a_b)
inter_a_b = set.intersection(set_a, set_b)
print(inter_a_b)
print(set_a-set_b)
```

```python
my_set_dict = {"a":set_a, "b":set_b,"c":set_c, "d":set_d }
venn(my_set_dict)
plt.savefig("my_venn.png", dpi = 1000)
```
