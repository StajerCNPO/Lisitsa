```python
import matplotlib.pyplot as plt
from venn import venn
```

```python
a = [1, 3, 5, 10, 22]
b = [2, 3, 8, 10, 44]
c = [5,6]
d = [1, 16, 33]
e = [2, 9, 44]

mn_a = set(a)
mn_b = set(b)
mn_c = set(c)
mn_d = set(d)
mn_e = set(e)
```

```python
uni_a_b = set.union(mn_a, mn_b)
print(uni_a_b)
inter_a_b = set.intersection(mn_a, mn_b)
print(inter_a_b)
print(mn_a - mn_b)
```

```python
my_dict = {"a":mn_a,"b":mn_b,"c":mn_c,"d":mn_d,"e":mn_e}
venn(my_dict)
plt.savefig("flower.png", dpi = 400)
```
